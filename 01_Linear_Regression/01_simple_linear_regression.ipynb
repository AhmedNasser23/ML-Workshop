```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Sample Data

This sample data is from ISLR. It displays sales (in thousands of units) for a particular product as a function of advertising budgets (in thousands of dollars) for TV, radio, and newspaper media.

```python
df = pd.read_csv('Datasets/Advertising.csv')
```

```python
df.head()
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df.head()
```

**Is there a relationship between *total* advertising spend and *sales*?**

```python
sns.scatterplot(data=df, x='total_spend', y='sales')
```

## Least Squares Line

```python
sns.regplot(data=df, x='total_spend', y='sales') # Will display line of best fit
```

```python
X = df['total_spend'] # feature
y = df['sales'] # target
```

### What actually this function doing is finding the best fit line, it calculates the parameters B1 and B0 by using ordinary least squares method
* B1 = pearson correlation coefficient * (std(y) / std(X))  
* B0 = mean(y) - B1 * mean(X)

```python
# y = mx+b
# y = B1x+B0
print(f"The parameters are: {np.polyfit(X,y,deg=1)}") 

# Pearson Correlation Coefficient
correlation_matrix = np.corrcoef(X, y)
pearson_correlation = correlation_matrix[0, 1]
print(f"Pearson Correlation Coefficient: {pearson_correlation}")
```

```python
potential_spend = np.linspace(0, 500, 100) # generate 100 values between 0 and 500
predicted_sales = 0.04868788*potential_spend + 4.24302822
plt.plot(potential_spend, predicted_sales, color='red')
plt.scatter(X, y)
plt.xlabel('Total Advertising Spend')
plt.ylabel('Sales')
plt.show()
```

```python
spend = 200
predicted_sale = 0.04868788*spend + 4.24302822
print(f"Predicted sales for advertising spend of ${spend}: {predicted_sale}")
```
